```python
#This code display a table having product’s ID, price list, sale price, and the discount columns, it then creates an empty data to store selected
#pricing data, and it checks if there any invalid numbers and replace it with nan

verified_product_df = (
    new_product_df[["Uniq_id", "list_price", "sale_price"]]
    .rename(columns={'list_price': 'List Price', 'sale_price': 'Sale Price'})
    .assign(Discount=lambda x: pd.to_numeric(x['List Price'], errors='coerce') -
     pd.to_numeric(x['Sale Price'], errors='coerce'))
    .dropna()
)
print(verified_product_df.head(2))
```

```python
#Displayed graph showing the price range of the products

# Convert 'Sale Price' to numeric values
verified_product_df['Sale Price']= pd.to_numeric(verified_product_df['Sale Price'], errors='coerce')

#state the edge of the bin inline with the price distribution
bin = [0, 50, 100, 200, 500, 18000]
labels = ['Below 50', '50-100', '100-200', '200-500', '500 - 18000']
verified_product_df['Price Range'] = pd.cut(
    verified_product_df['Sale Price'],
    bins=bin,
    labels=labels, right=False)

# Plot a clear histogram
plt.figure(figsize=(9, 6))
sns.countplot(data=verified_product_df, x='Price Range', color='skyblue')

# Add chart details
plt.title("Distribution of Price Range", fontsize=14)
plt.xlabel('Sale Price Range')
plt.ylabel('Number of Products')
plt.grid(axis='y', alpha=0.7)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# This add the reviewers_json data frame and reviews_cs data information and add up the user's reviews and
# then combine them with their reviewed products and then equate the number of time eacg of the product gets
# reviewed by the customers

#Combining reviewers_json data and reviews_csv data together
combine_the_reviews_data = pd.merge(reviews_csv_df, reviewers_json_df,  on='Username')

#Grouping by the 'Username' and then aggregate the columns
combine_summary_reviews = (
    combine_the_reviews_data.groupby('Username')
    .agg({
        'Uniq_id': lambda x: list(x.unique()),
        'Score': list,
        'Review': lambda x: ', '.join(x.unique()),
        'DOB': lambda x: ', '.join(x.unique()),
        'State': lambda x: list(x.unique())
    })
    .reset_index()
)

#Combine with reviewers dataframe to add the Reviewed products
combine_summary_reviews = pd.merge(
    combine_summary_reviews,
    reviewers_json_df[['Username', 'Reviewed']],
    on='Username'
)

# Combine 'Uniq_id' and 'Reviewed' into one list, then drop 'Uniq_id'
combine_summary_reviews['Reviewed'] = combine_summary_reviews['Uniq_id'] + combine_summary_reviews['Reviewed']

# Count number of reviewed items per user
combine_summary_reviews['Reviewed Items By Count'] = combine_summary_reviews['Reviewed'].apply(len)

# Display final summary
print(combine_summary_reviews.head(2))
```

```python
#The overall number of items reviewed by customers in each state

#Calculate the overall reviewed items in each of the states
state_review_totals = (
    combine_summary_reviews.explode('State')
    .groupby('State')['Reviewed']
    #Count the overall items reviewed within each state
    .apply(lambda x: sum(len(i) for i in x))  
    .reset_index(name='Total Reviewed Items')
)
#Print a vertical bar chart
plt.figure(figsize=(12, 6))
sns.barplot(
    data=state_review_totals.sort_values('Total Reviewed Items', ascending=False),
    x='State',
    y='Total Reviewed Items',
    color='steelblue'
)
plt.title("total number of items reviewed across the states by customers", fontsize=14)
plt.xlabel("State")
plt.ylabel("Total Reviewed Items")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

```python
#Checking for the most frequently used words by the viewers. I will import and use the wordcloud to review the most used words by reviewers
from sklearn.feature_extraction.text import CountVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

#Combine the reviews into single text block for clarity
all_reviews = ' '.join(combine_summary_reviews['Review'])
#Create CounterVectorize to convert words to numerical counts and using stopwords to auto remove any common words
vectorizer = CountVectorizer(stop_words=list(ENGLISH_STOP_WORDS))
#Fitting the texts in vectorizer and convert to word count
word_bag = vectorizer.fit_transform([all_reviews])
#Fetch the feature names
feature_names = vectorizer.get_feature_names_out()
#Convert the word bag to a dictionary for WordCloud
word_frequencies = dict(zip(feature_names, word_bag.toarray()[0]))
#Build a WordCloud with frequency dict
wordcloud = WordCloud(
    width=700, 
    height=400, 
    background_color='white'
).generate_from_frequencies(word_frequencies)

plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title("Customers Frequently Used Words Reviews")
plt.axis('off')
plt.show()
```
